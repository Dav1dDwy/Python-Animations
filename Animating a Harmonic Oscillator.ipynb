```python
import matplotlib.pyplot as plt
import numpy as np
% matplotlib inline 
```

```python
#Generate some time values

time = np.arange(0,101,1)


#define function for an arbitrary harmonic oscillator

def hosc(t,w):
    x = np.cos(w*t)
    return x

#determine x values using time values and value of 0.5 for w

output = []

for i in time:
    output.append(hosc(i,0.5))
    
#Plot a static graph of x against time

output = np.array(output)

plt.plot(time,output,'o-')

plt.title("Harmonic oscilllator for $\omega_0 = 0.5$")
plt.xlabel("time")
plt.ylabel("position")
```

```python
#Generating a animation over time of the motion of the oscillator

from matplotlib.animation import FuncAnimation
from matplotlib import animation
```

```python
#Generate figure and declare the axes

fig = plt.figure()
ax = plt.axes(xlim = (-10,110),ylim = (-1.5,1.5))
line, = ax.plot([], [], lw = 2)       #line being the object to be changed for the animation


def init():
    line.set_data([],[])
    return line,

def animate(i):
    t = time
    w = 0.5
    position = np.cos(w*(t-i))
    line.set_data(t,position)       #Giving line the values of time and position in an order to 
    return line,                 #be animated
    

#making the animation

anim = animation.FuncAnimation(fig,animate, frames = 200,
                               init_func = init, blit = True)



#Save the animation as an mp4 file


writer = animation.FFMpegWriter(fps=30, metadata=dict(artist='Me'), bitrate=1800)

animation.Animation.save(anim, "Oscillator.mp4" )
plt.show()
```
